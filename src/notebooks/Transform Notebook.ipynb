```python
import pandas  as pd 
pd.set_option('display.max_columns', None)
import numpy as np

df = pd.read_csv('..\data\Openrent.csv')
```

```python
df['Monthly Rent'] = df['Rent PCM'].apply(lambda x: x.split('.')[0])
df.drop('Rent PCM',axis=1, inplace= True)
```

```python
def to_text(number):
    if 10 <= number % 100 <= 20:
        suffix = 'th'
    else:
        suffix = {1: 'st', 2: 'nd', 3: 'rd'}.get(number % 10, 'th')
    return str(number) + suffix +' floor'

df['Floor'] = np.random.randint(1, 20, size=df.shape[0])
df['Floor'] = df["Floor"].apply(to_text)
```

```python
df.columns = [col.strip(' ') for col in df.columns]
df.columns
```

```python
df.drop('Deposit', axis = 1, inplace =  True )
new_cols =  ['title', 'description', 'Monthly Rent', 'Floor','Bills Included', 'Broadband',
       'Student Friendly', 'Families Allowed', 'Pets Allowed',
       'Smokers Allowed', 'DSS/LHA Covers Rent', 'Available From',
       'Minimum Tenancy',  'Garden', 'Parking',
       'Fireplace', 'Furnishing', 'EPC Rating', 'station_distances',
       'DSS Income Accepted', 'Maximum Tenancy', 'Students Only','EPC Not Required']


df = df[new_cols]
df
```
