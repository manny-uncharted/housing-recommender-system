```python
df1 = pd.read_csv('/content/drive/MyDrive/rental_service/zoopla data/dataset_zoopla-scraper_2023-07-16_09-04-20-564.csv')
df2 = pd.read_csv('/content/drive/MyDrive/rental_service/zoopla data/dataset_zoopla-scraper_2023-07-16_09-04-59-621.csv')

df = pd.concat([df1,df2])
df
```

```python
use_cols = ['address', 'agent', 'agentPhone', 'category', 'coordinates/latitude','coordinates/longitude', 'countyArea', 'description', 'features/0',
       'features/1', 'features/2', 'features/3', 'features/4', 'features/5','features/6', 'features/7', 'features/8', 'floor', 'id', 'incode',
       'listingsCategory', 'numBathrooms', 'numBedrooms', 'numLivingRooms',
       'outcode', 'pointsOfInterest/0/distance', 'pointsOfInterest/0/title', 'pointsOfInterest/1/distance', 'pointsOfInterest/1/title',
       'pointsOfInterest/2/distance', 'pointsOfInterest/2/title',
       'pointsOfInterest/3/distance', 'pointsOfInterest/3/title', 'postalCode',
       'price', 'priceMax', 'priceMin', 'propertyType', 'publishedOn',
       'region', 'sizeSqFeet', 'title', 'type']

new_df = df[use_cols]
```

```python
import re
new_df['price'] = new_df['price'].apply(lambda x: eval(re.sub(r"\D", "", x)))
new_df
```

```python
cat_cols = ['category','countyArea','listingsCategory','propertyType']

new_df.fillna(-999, inplace = True)
new_df = pd.get_dummies(new_df, columns = cat_cols)
```

```python
drop_cols = ['title','address', 'agent', 'agentPhone', 'description', 'features/0',
       'features/1', 'features/2', 'features/3', 'features/4', 'features/5', 'postalCode',
       'features/6', 'features/7', 'features/8', 'floor', 'id', 'incode','outcode', 'pointsOfInterest/0/distance',
      'pointsOfInterest/0/title', 'pointsOfInterest/1/distance', 'pointsOfInterest/1/title','pointsOfInterest/3/title',
       'pointsOfInterest/2/distance', 'pointsOfInterest/2/title','pointsOfInterest/3/distance', 'pointsOfInterest/3/title'
       'postalCode', 'price', 'priceMax', 'priceMin',  'publishedOn','region', 'sizeSqFeet', 'title', 'type']

feat_cols = new_df.columns.difference(drop_cols)
```

```python
X = new_df[feat_cols]

y = new_df['price']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,shuffle=True)
```

```python
!pip install catboost -qq
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42,max_depth = 6)
rf.fit(X_train,y_train)
preds = rf.predict(X_test)
print(f'Model has a mean absolute error of {mean_absolute_error(y_test,preds)} and RMSE of {mean_squared_error(y_test,preds,squared= False)}')
print(f'Model has a r2 score of {r2_score(y_test,preds)}')


plt.scatter(y_test,preds)
plt.title('Scatter Plot Showing relation between actual and predicted values')
plt.show()
```

```python
xgb = XGBRegressor(n_estimators=100, random_state=42,max_depth = 6,learning_rate=0.1)
xgb.fit(X_train,y_train)
preds = xgb.predict(X_test)
print(f'Model has a mean absolute error of {mean_absolute_error(y_test,preds)} and RMSE of {mean_squared_error(y_test,preds,squared= False)}')
print(f'Model has a r2 score of {r2_score(y_test,preds)}')

plt.scatter(y_test,preds)
plt.title('Scatter Plot Showing relation between actual and predicted values')
plt.show()
```

```python
cat=CatBoostRegressor(learning_rate= 0.02,iterations=650,depth= 6,subsample=0.8,
                      bootstrap_type='Bernoulli',random_seed= 12,task_type='CPU',verbose= 0 )

cat.fit(X_train,y_train)
preds = cat.predict(X_test)
print(f'Model has a mean absolute error of {mean_absolute_error(y_test,preds)} and RMSE of {mean_squared_error(y_test,preds,squared= False)}')
print(f'Model has a r2 score of {r2_score(y_test,preds)}')

plt.scatter(y_test,preds)
plt.title('Scatter Plot Showing relation between actual and predicted values')
plt.show()
```
