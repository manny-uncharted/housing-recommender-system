```python
import pandas  as pd 
pd.set_option('display.max_columns', None)
import numpy as np

df1 = pd.read_csv('Rentola-1.csv')
df2 = pd.read_csv('Rentola 2.csv')

df = pd.concat([df1,df2])
```

```python
df.columns = [col.replace(':','') for col in df.columns]
df
```

```python
df['Price'] = df['Price'].apply(lambda x: int(x.replace('€','')))
df
```

```python
use_df = df[['Property type','City','Furnished','Bedrooms','Price','latitude','longitude','Bathrooms']]

cat_cols = ['Property type', 'City','Furnished']
use_df = pd.get_dummies(use_df, columns = cat_cols)
use_df
```

```python
use_df.fillna(-999, inplace = True)
use_df
```

```python
X = use_df.drop('Price', axis = 1)
y = use_df['Price']
```

```python
from sklearn.model_selection import train_test_split

# Set the test size to 30% of the data and enable shuffling
test_size = 0.30
shuffle = True

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, shuffle=shuffle, random_state = 12)
```

```python
from catboost import CatBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
import random
import os
```

```python
# Dictionary to store the different metric results
mean_squared_errors_results = {}
mean_absolute_errors_results = {}
root_mean_squared_errors_results = {}
r2_scores_results = {}

def compute_metrics(y_test, preds, model_name):

    mse = mean_squared_error(y_test, preds)
    mean_squared_errors_results[model_name] = mse
    print(f'Mean Squared Error: {mse}')

    rmse = mean_squared_error(y_test, preds, squared=False)
    root_mean_squared_errors_results[model_name] = rmse
    print(f'Root Mean Squared Error: {rmse}')

    mae = mean_absolute_error(y_test, preds)
    mean_absolute_errors_results[model_name] = mae
    print(f'Mean Absolute Error: {mae}')

    r_squared = r2_score(y_test, preds)
    r2_scores_results[model_name] = r_squared
    print(f'R2 Score: {r_squared}')
```

```python
rf = RandomForestRegressor(n_estimators=500, random_state=42,max_depth = 8)
rf.fit(X_train,y_train)
preds = rf.predict(X_test)
# evaluating the model on the test set
compute_metrics(y_test, preds, 'rf_model')
```

```python
xgb=XGBRegressor(n_estimators1=550 , max_depth=4, learning_rate=0.08, colsample_bytree=0.6,verbosity = 0,
                  random_state=12,subsample=0.8)

xgb.fit(X_train,y_train)
preds = xgb.predict(X_test)
# evaluating the model on the test set
compute_metrics(y_test, preds, 'xgb_player_features')
```

```python
cat=CatBoostRegressor(learning_rate= 0.08,iterations=650,depth= 6,subsample=0.8,
                      bootstrap_type='Bernoulli',random_seed= 12,task_type='CPU',verbose= 0 )

cat.fit(X_train,y_train)
preds = cat.predict(X_test)
# evaluating the model on the test set
compute_metrics(y_test, preds, 'cat_player_features')
```

```python
cat.fit(X,y)

cat.save_model('catboost_model')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns 
def plot_feature_importance(importance,names,model_type,n=40):

    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)

    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    fi_df=fi_df.nlargest(n, ['feature_importance'], keep='first')
    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)

    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + ' FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')

plot_feature_importance(cat.get_feature_importance(),X.columns,'CATBOOST')
```
