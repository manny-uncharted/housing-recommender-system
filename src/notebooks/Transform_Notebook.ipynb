```python
import pandas  as pd
pd.set_option('display.max_columns', None)
import numpy as np

df = pd.read_csv('Openrent.csv')
```

```python
df['Monthly Rent'] = df['Rent PCM'].apply(lambda x: x.split('.')[0])
df.drop('Rent PCM',axis=1, inplace= True)
```

```python
def to_text(number):
    if 10 <= number % 100 <= 20:
        suffix = 'th'
    else:
        suffix = {1: 'st', 2: 'nd', 3: 'rd'}.get(number % 10, 'th')
    return str(number) + suffix +' floor'

df['Floor'] = np.random.randint(1, 20, size=df.shape[0])
df['Floor'] = df["Floor"].apply(to_text)
```

```python
df2 = pd.read_csv('Rentola-1.csv')
df2
```

```python

df2.columns = [col.replace(':','') for col in df2.columns]
df2[['Elevator', 'Washing machine', 'Dish washer', 'Pets allowed', 'Garage','Terrace', 'Swimming pool', 'Balcony']].fillna('No', inplace = True)

df2.rename(columns = {'Price':'Monthly Rent'},inplace = True)
df2
```

```python
df2['Floor'] = np.random.randint(1, 20, size=df2.shape[0])
df2['Floor'] = df2["Floor"].apply(to_text)
```

```python
df3 = pd.read_csv('rightmove.csv',lineterminator='\n')
df3.rename(columns = {'Price':'Monthly Rent'},inplace = True)
df3['Deposit'] = df3['Deposit'].apply(lambda x:x.split('A deposit')[0])
df3['Floor'] = np.random.randint(1, 20, size=df2.shape[0])
df3['Floor'] = df3["Floor"].apply(to_text)
df3
```

```python
num = 0
with open("All properties.txt", "w", encoding='utf-8') as text_file:
    for data in [df,df2,df3]:
        for i in data.iterrows():
            num+=1
            text_file.write(f'{num+1}. ')
            line = '.\n'.join([f'{y}: {z}' for y,z in zip(i[1].index, i[1].values)])
            text_file.write(line)
            text_file.write('\n')
```
